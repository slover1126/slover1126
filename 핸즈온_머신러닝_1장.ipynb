**1.1 머신러닝이란**

  머신러닝이란 명시적인 프로그래밍 없이 컴퓨터가 학습하는 능력을 갖추게 하는 연구 분야이다.

  -용어 정리

 **Training Set**: 시스템이 학습하는 데 사용하는 샘플

  **Training instance/Sample**:각각의 훈련 데이터

  **Model**: 학습하고 예측을 만드는 부분

 **1.2 머신러닝을 이용하는 이유**

 1. 기존에 알고리즘/프로그램이 있던 경우라도 지속적인 유지보수가 필요한 경우

 2. 너무 복잡하거나 알려진 알고리즘이 없는 경우

 3. 새로운 규칙을 찾는 경우(데이터 마이닝)

 **1.4 머신러닝 시스템의 종류**

 머신러닝의 종류는 크게 훈련 지도 방식, 실시간 점진적 학습 여부, 사례와 모델 기반으로 나눌 수 있으며 배타적인것은 아님.

 **1.4.1 훈련 지도 방식**

1.  지도 학습(훈련 데이터에 **레이블**이라는 답을 포함하여 학습)

    분류와 회귀가 있으며 분류는 클래스를 나누고, 회귀는 수치를 예측하는것이 주요 목표임.


2. 비지도 학습(훈련 데이터에 레이블 없이 학습이 진행)
    
    군집분석, 시각화, 차원축소, 이상치 탐지,연관규칙 학습등이 있음

3. 준지도 학습(훈련 데이터에 레이블이 일부만 있는 경우)

    대부분 지도학습과 비지도 학습의 조합으로 이루어짐

4. 자기지도 학습(레이블이 없는 데이터 셋에서 레이블이 완전히 부여된 데이터셋을 생성)



5. 강화 학습(**학습하는 시스템이 보상/벌점을 통해 정책이라는 최상의 전략을 스스로 학습함**)

    보행 로봇을 만들기 위해 주로 사용되고, 알파고에도 사용되었음

 **1.4.2 배치 학습과 온라인 학습**
    
1. 배치 학습(오프라인 학습)

    *   가용한 모든 데이터를 이용하여 학습을 진행하는 형태
    *   시간이 지날수록 모델의 성능이 감소하는 현상 발생(모델 부패/데이터 드리프트)
    * 계속 학습을 시켜야 하기 때문에 많은 컴퓨팅 자원을 필요로 하고, 변동성이 큰 데이터를 다룬다면 적합하지 않은 방식임



2. 온라인 학습

    *   데이터를 한개씩, 혹은 **미니 배치** 를 활용해서 계속 학습을 진행하는 형태
    *   컴퓨팅 자원이 제한되어있거나 변동성이 큰 데이터를 다루는 경우에도 적합
    * 전체 데이터를 모두 학습할 때까지 진행됨

**학습률의 문제**

학습률:변화하는 데이터에 얼마나 빠르게 적응할 지에 대한 민감도

너무 빠르게 적응한다면 과거 데이터를 잊어버리고, 너무 느리게 적응한다면 데이터가 모델에 영향을 주지 못하는 문제가 생김

**1.4.3 사례 기반 학습과 모델 기반 학습**


1.  사례 기반 학습



    *   시스템이 훈련 샘플을 기억하는 방법으로 학습


    *   유사도(주로 공통으로 가진 원소로 측정)로 분류


2.  모델 기반 학습

    *   모델을 만들어 **예측**을 시행
